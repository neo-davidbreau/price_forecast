```python
import mlflow, pandas as pd
from prophet import Prophet
```

```python
mlflow.start_run(run_name="prophet_forc")
```

```python
data = pd.read_csv('data/gold_price_data.csv')
```

```python
model = Prophet()
```

```python
data = data.rename(columns={'Date': 'ds', 'Value': 'y'})
```

```python
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365)  # Prvision pour un an
forecast = model.predict(future)

# Afficher les prvisions
fig = model.plot(forecast)
```
